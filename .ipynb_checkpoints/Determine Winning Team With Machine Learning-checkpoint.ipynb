```python
import pandas as pd
```

```python
df = pd.read_csv('teams/teams.csv')
df.describe()
```

```python
df[['teamScore', 'opptTeamScore']].tail(10)
```

```python
df.info()
```

```python
df['gameKey'] = df['gameKey'].astype(int)
df['offenseScore'] = df['offenseScore'].astype(int)
df['defenseScore'] = df['defenseScore'].astype(int)
df['teamScore'] = df['teamScore'].astype(int)
df['opptOffenseScore'] = df['opptOffenseScore'].astype(int)
df['opptDefenseScore'] = df['opptDefenseScore'].astype(int)
df['opptTeamScore'] = df['opptTeamScore'].astype(int)
df['points'] = df['points'].astype(int)
df['pointsAgainst'] = df['pointsAgainst'].astype(int)
df.info()
```

```python
df['eloDiff'] = df.teamScore - df.opptTeamScore
df['eloDiff'].tail(10)
```

```python
df[['teamScore','opptTeamScore']].tail(10)
```

```python
df['pointsDiff'] = df.points - df.pointsAgainst
df[['points', 'pointsAgainst', 'pointsDiff', 'eloDiff']].tail(5)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

df.hist(bins=50, figsize=(20,15))
plt.savefig("attribute_histogram_plots")
plt.show()
```

```python
df.plot(kind="scatter", x="win", y="eloDiff", alpha=0.2)
plt.savefig('winmap.png')
```

```python
df.plot(kind='scatter', x='win', y='eloDiff', alpha=0.4, figsize=(10,7),
        c='pointsDiff', cmap=plt.get_cmap("jet"), colorbar=True, sharex=False)
```

```python
new_df = df.drop(['points', 'pointsDiff', 'pointsAgainst'], axis = 1)
corr_matrix = new_df.corr()
corr_matrix['win'].sort_values(ascending=False)
```

```python
from pandas.tools.plotting import scatter_matrix

attributes = ['eloDiff', 'teamScore', 'scoreAvg', 'home', 'totalAvg']
scatter_matrix(new_df[attributes], figsize=(12,8))
plt.savefig('matrix.png')
```
