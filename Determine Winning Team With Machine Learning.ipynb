```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('teams/teams.csv')
df.describe()
```

```python
df[['teamScore', 'opptTeamScore']].tail(10)
```

```python
df.info()
```

```python
df['gameKey'] = df['gameKey'].astype(int)
df['offenseScore'] = df['offenseScore'].astype(int)
df['defenseScore'] = df['defenseScore'].astype(int)
df['teamScore'] = df['teamScore'].astype(int)
df['opptOffenseScore'] = df['opptOffenseScore'].astype(int)
df['opptDefenseScore'] = df['opptDefenseScore'].astype(int)
df['opptTeamScore'] = df['opptTeamScore'].astype(int)
df['points'] = df['points'].astype(int)
df['pointsAgainst'] = df['pointsAgainst'].astype(int)
df.info()
```

```python
df['eloDiff'] = df.teamScore - df.opptTeamScore
df['eloDiff'].tail(10)
```

```python
df['totalDiff'] = df.totalAvg - df.opptTotalAvg
df.totalDiff.tail(10)
```

```python
df['passDiff'] = df.passAvg - df.opptPassAvg
df.passDiff.tail(10)
```

```python
df['rushDiff'] = df.rushAvg - df.opptRushAvg
df.rushDiff.tail(10)
```

```python
df[['teamScore','opptTeamScore']].tail(10)
```

```python
df['pointsDiff'] = df.points - df.pointsAgainst
df[['points', 'pointsAgainst', 'pointsDiff', 'eloDiff']].tail(5)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

df.hist(bins=50, figsize=(20,15))
plt.savefig("attribute_histogram_plots")
plt.show()
```

```python
df['pointPerYard'] = 0
for team in df.team.unique():   
    teamAvg = []
    for col, row in df.ix[df.team == team].iterrows():
        try:
            pointPerYard = row.points / row.totalAvg
            teamAvg.append(pointPerYard)
            avg = np.mean(teamAvg)
            row.pointPerYard = avg
        except ZeroDivisionError:
            pass
df.pointPerYard.tail(10)
```
